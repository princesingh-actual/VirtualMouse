```python
path = r'C:\Users\Prince\Desktop\team_light\Lift_operation\Dataset\\'
#path for dataset
import cv2
import xlsxwriter
import glob
import mediapipe as mp
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(
    min_detection_confidence=0.45, min_tracking_confidence=0.45)



expenses = []
for i in range(21):
    expenses.append(str(i)+"X")
    expenses.append(str(i)+"Y")
    expenses.append(str(i)+"Z")
expenses.append("Target")

workbook = xlsxwriter.Workbook('Excel_lift.xlsx')
worksheet = workbook.add_worksheet()
row = 0
col = 0
for item in (expenses):
    worksheet.write(row, col,item)
    col += 1

    
    
row=1
col=0
patth=path
alpha="0123456789DOU$"
targetnum=0
for ii in alpha:
    nohand=1
    path=patth+ii+"\*.jpg"
    print(path)
    path = glob.glob(path)
    print(ii)
    print(targetnum)
    for img in path:
        frame = cv2.imread(img)
        image = cv2.cvtColor(cv2.flip(frame, 1), cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        results = hands.process(image)

        l=[]
        if results.multi_hand_landmarks:
            for hand_landmarks in results.multi_hand_landmarks:
                land = hand_landmarks
                val = land.landmark
                first=val[0]
                for j in val:
                    l.append([j.x-first.x,j.y-first.y,j.z-first.z])
                break
        else:
            print(nohand)
            nohand+=1
            continue
        col=0
        for item in l:
            worksheet.write(row, col,item[0])
            col+=1
            worksheet.write(row, col,item[1])
            col+=1
            worksheet.write(row, col,item[2])
            col+=1
        worksheet.write(row, col,targetnum)
        row+=1
    targetnum+=1
workbook.close()
```
